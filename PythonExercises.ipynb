```python
import ast

# Safely evaluate the string 'a' into a Python list
parsed_list = ast.literal_eval(a)

# Convert each element in the parsed list to an integer
int_list = [int(item) for item in parsed_list]

print(f"Converted list: {int_list}")
display(int_list)
```

```python
int_list.sort(reverse=True)
print(int_list)
```
